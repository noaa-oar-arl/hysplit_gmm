```python
def makehist(mass, ahash,name, label=''):
    #sns.set_style('whitegrid')
    sns.set(font_scale=1.5)
    sns.set_style('whitegrid')
    minval = 1e-2
    mass2 = par2conc.threshold(mass,tval=minval,tp='linear',fillna=False)
    xr.plot.hist(mass2,**ahash)
    plt.xlabel('Mass Loading (g m$^{-2}$)')
    plt.ylabel('Number')
    plt.title('')
    plt.text(2.7,500,label)
    plt.tight_layout()
    plt.savefig(name + '.png')
```

```python
def fixplot(xin, fname, loc=145,label=''):
    sns.set(font_scale=1.5)
    sns.set_style('whitegrid')
    multstr='C'
    if multstr=='A':
        smin = 0.01
        smax = 0.2
        smin = 0.005
    else:
        smin = 0.01
        smax = 3
        
    if loc==145:
        xrange = [48,55]
        yrange = [7,15]
        tx =[54,14]
    else:
        xrange = [40,55]
        yrange = [1,15]
        smax = 1.0 *10
        tx = [53.5,13.5]
    cmap = 'viridis'              
    chash = {'label': 'mg m$^{-3}$'}
    xin.plot.pcolormesh(x='latitude',y='z',
                        cbar_kwargs=chash,
                        cmap=cmap,
                        vmin=smin,
                        vmax=smax)
    plt.xlim(xrange[0],xrange[1])
    plt.ylim(yrange[0],yrange[1])
    plt.title('')
    plt.text(tx[0],tx[1],label)
    plt.ylabel('height (km)')
    plt.xlabel('Latitude')
    plt.tight_layout()
    plt.savefig(fname + '.png')          
        

    
```

```python
from utilhysplit import concutils
def slicefit(conc,mult,dx=5,dz=10,loc=-145,thresh=None):
    if dx>0:
        conc_coarsen = conc.coarsen(x=dx,boundary='trim').mean()
        conc_coarsen = conc_coarsen.coarsen(y=dx,boundary='trim').mean()
    else:
        conc_coarsen = conc
    if dz>0:
        conc_coarsen = conc_coarsen.coarsen(z=dz,boundary='trim').mean()
    conc2 = concutils.xslice(conc_coarsen*mult*1000, loc)
    if thresh:
        conc2 = par2conc.threshold(conc2,tval=thresh,tp='linear',fillna=False)
    return conc2
```

```python
def plotpart(df,conc,loc=-145,figname='junk',ms=[5,2],label=''):
    
    sns.set()
    sns.set(font_scale=1.5)
    sns.set_style('whitegrid')
    htmult = 1/1000.0
    cmap = 'viridis'
    if loc == -145:
        tx = [54,14]
        xrange = [48,55]
        yrange = [7,15]
        levels = [0.001,0.005,0.01,0.05,0.1]
        levels = np.array(levels)*10
    elif loc == -151:
        xrange = [40,55]
        yrange = [1,15]
        levels = [0.001,0.01,0.1,0.2,0.5]
        levels = np.array(levels)*10
        tx = [53.5,13.5]
    dd=0.125
    temp = df[df.lon < (loc + dd)]
    temp = temp[temp.lon > (loc - dd)]
    dd=0.5
    temp2 = df[df.lon < (loc + dd)]
    temp2 = temp2[temp2.lon > (loc - dd)]
    plt.plot(list(temp2.lat), list(temp2.ht*htmult), 'ro',MarkerSize=ms[1])
    plt.plot(list(temp.lat), list(temp.ht*htmult), 'ko',MarkerSize=ms[0])
    plt.xlabel('Latitude (degrees)')
    plt.ylabel('Height (km)')
    plt.text(tx[0],tx[1],label)
    plt.xlim(xrange[0],xrange[1])
    plt.ylim(yrange[0],yrange[1])
    #levels = [0.001,0.005,0.01,0.05,0.1]
    #levels = [0.001,0.005,0.01,0.05,0.1]
    cb = xr.plot.contour(conc, x='latitude',y='z',levels=levels,cmap=cmap,linewidths=3)
    plt.colorbar(cb)
    plt.title('')

    plt.xlabel('Latitude')
    plt.ylabel('Height (km)')
    plt.tight_layout()
    plt.savefig(figname + '.png')
    
```
